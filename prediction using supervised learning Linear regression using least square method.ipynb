```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)

#reading Data
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
print(data.shape)
data.head()
```

```python
data.head(10)
```

```python
#relationship between two variables
#collecting X and Y
X=data['Hours'].values
Y=data['Scores'].values
```

```python
#Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

#total number of values
m= len(X)


#Using the formula to calculate b1 and b2
numer=0
denom = 0
for i in range(m):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom +=(X[i]-mean_x)**2
b1=numer/denom
b0= mean_y - (b1*mean_x)
#print coefficients
print(b1,b0)
```

```python
#plotting values and Regression Line

max_x = np.max(X)
min_x= np.min(X)

#calculating line vaue x and y
x=np.linspace(min_x,max_x,10)
y=b0+b1*x
#plotting Line
plt.plot(x,y,color ='#58b970',label ='Regression line')

#plotting scatter Points
plt.scatter(X,Y,color ='#ef5423',label ='Scatter plot')

plt.xlabel ('Hours')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# r square method and its value
ss_t=0 # total sum of square
ss_r=0  #total sum of square residuals

for i in range(m):
    y_pred = b0+b1*X[i]
    ss_t +=(Y[i]- mean_y)**2
    ss_r +=(Y[i]-y_pred)**2
r2 = 1-(ss_r/ss_t) # rsquare formula
print(r2)
    
```

```python
# NOW USING SCKIKIT LEARN LIBRABRY
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#cannot use Rank 1 matrix in scikit learn
X=X.reshape((m,1))
#creating Model
reg=LinearRegression()

#fitting training data
reg = reg.fit(X,Y)
# Y prediction

Y_pred =reg.predict(X)

#calculation R2 Score
r2_score=reg.score(X,Y)
print(r2_score)
```
